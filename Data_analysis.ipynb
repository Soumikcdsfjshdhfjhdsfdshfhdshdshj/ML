#### Univariate , Bivariate And Multivariate

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris = sns.load_dataset("iris")
```

```python
iris
```

### Univariate

```python
iris_setosa=iris.loc[iris['species']=='setosa']
```

```python
iris_virginica=iris.loc[iris['species']=='virginica']
iris_versicolor=iris.loc[iris['species']=='versicolor']
```

```python
plt.plot(iris_setosa['petal_length'], np.zeros_like(iris_setosa['sepal_length']),'o')
plt.plot(iris_virginica['petal_length'], np.zeros_like(iris_virginica['sepal_length']),'o')
plt.plot(iris_versicolor['petal_length'], np.zeros_like(iris_versicolor['sepal_length']),'o')
plt.xlabel('petal_length')
```

### Bivariate

```python
sns.FacetGrid(iris,hue="species").map(plt.scatter,"sepal_length","sepal_width").add_legend();
plt.show()
```

### Multivariate 

```python
sns.pairplot(iris, hue="species", size =5)
```
